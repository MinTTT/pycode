```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

image = cv.imread('FM15-M13GFP-2.png', 0)
image_edge_detect = cv.GaussianBlur(image,(7,7), 0)
edge = cv.Canny(image_edge_detect,18, 20)
edge = cv.dilate(edge, None, iterations=1)
edge = cv.erode(edge, None, iterations=1)
edge, contours, hierarchy = cv.findContours(edge, cv.RETR_TREE, cv.CHAIN_APPROX_NONE)

filteredcnt = [ i for i in contours if cv.contourArea(i)>160]

x,y,w,h = cv.boundingRect(filteredcnt[0])
img = cv.rectangle(image.copy(),(x,y),(x+w,y+h),(0,255,255),4)

#print(np.array(filteredcnt).shape)

plt.figure(figsize=(16,35))
plt.subplot(121)
plt.imshow(image_edge_detect, cmap = 'gray')
plt.subplot(122)
plt.imshow(img)
plt.show()
```

### Note
* cv.imread(path, 0) use parameter '0' to read a image as grayscale
